### 1.Temperature Conversion

````python
def convert_temp(temp,unit):
    """ This function converts temp b/w celsius and fahrenheit"""
    if unit == 'C':
        return temp*9/5 + 32 # celsius to fahrenheit
    elif unit == 'F':
        return (temp-32)*5/9 # fahrenheit to celsius
    else:
        return None

````

````python
print(convert_temp(32,'C')) # celsius to fahrenheit
print(convert_temp(77,'F')) # fahrenheit to celsius
````

### 2.Password Strength Checker

````python
def check_password(password):
    """The function checks the password is strong or not """
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password ):
         return False
    if not any(char.islower() for char in password):
         return False
    if not any (char.isupper() for char in password):
         return False
    if not any (char in '!@#$%^&*()_+' for char in password):
        return False
    else:
        print('password is strong')
````

````python
check_password('ak@achyuth')
````

### 6. Function to read a file and count the frequency of each word

````python
def count_word_frequency(file_path):
    word_count = {}
    with open(file_path,'r') as file :
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip(',.!@#$%^&*()_+')
                word_count[word] = word_count.get(word,0)+1
    return word_count

file_path = 'sample.txt'
word_frequency = count_word_frequency(file_path)
print(word_frequency)
````

## Lambda

#### This function can have any number of arguments but only one expression, which is evaluated and returned.

## syntax = lambda arguments : expression

````python
# example
x = lambda a : a + 10
print(x(5))
````

````python
la = lambda x,y,z : x>y<z
print(la(6,3,4))   # x >y 6>3 , y<z 3<4 is True
````

````python
la = lambda x,y,z : x+y+z and x*y*z
print(la(6,3,4))  # it takes multiplicaton
````

 The issue is with the x + y + z and x * y * z expression inside the lambda function. Here's how it gets evaluated:
x + y + z is computed first. It results in a value, let's say v1.
v1 and x * y * z is then evaluated. 
The and operator returns the second operand (x * y * z) when the first one (v1) is a non-zero value (which is always the case for a valid sum of integers in this case).
As a result, the sum (x + y + z) is essentially ignored, and only the product (x * y * z) will be returned.

 Exceptions, on the other hand, occur during the execution of the program. *These errors are not detected by the parser but are raised during runtime *when a specific condition is encountered. Exceptions can be handled using try-except blocks

## How to handle exceptions 

Exceptions in Python can be handled using the try, except, else, and finally blocks. The try block contains the code that might raise an exception, and the except block handles the exception

#### try: This block contains the code that might raise an exception. It is the part of the code that needs to be monitored for potential errors.
#### except: If an exception occurs in the try block, the corresponding except block is executed. It handles and catches the exception, preventing the program from crashing.
#### else: This block is executed if no exceptions occur in the try block. It contains code that should run only when there are no exceptions.
#### finally: This block always gets executed, whether an exception occurs or not. It is useful for releasing resources or cleaning up, ensuring that certain actions are taken regardless of the outcome.

````python
# exceptions can be raised explicitly using the raise statement.
def divide_numbers(num1, num2):
    if num2 == 0:
        raise ValueError("Cannot divide by zero.")
    else:
        return num1 / num2

try:
    result = divide_numbers(10, 0)

except ValueError as e:
    print("Error:", e)

else:
    print("Result:", result)
````

### Creating a Custom Exception in Python:

#### In Python, you can create custom exceptions by defining a new class that inherits from the built-in Exception class or one of its subclasses. This allows you to raise and catch exceptions that are specific to your application or use case.

````python
class CustomError(Exception):
    def __init__(self, message="Custom error occurred"):
        self.message = message
        super().__init__(self.message)

def custom_function(value):
    try:
        if value < 0:
            raise CustomError("Negative value not allowed")

        # Perform some operations with the value
        result = 10 / value

    except CustomError as ce:
        print(f"CustomError: {ce}")

    else:
        print("No custom error occurred.")

    finally:
        print("Cleanup or finalization code here.")

# Example usage
custom_function(5)
custom_function(-2)
````

````python
 #The sys.exc_info() function in Python is used to get information about the most recent exception that occurred. 
#It returns a tuple containing information about the current exception, including the exception type, the exception instance, and the traceback.

import sys

def example_function():
    try:
        # Some code that may raise an exception
        x = 1 / 0

    except ZeroDivisionError as e:
        # Get information about the exception
        exc_type, exc_value, exc_traceback = sys.exc_info()

        # Print exception details
        print(f"Exception Type: {exc_type}")
        print(f"Exception Value: {exc_value}")
        print(f"Exception Traceback: {exc_traceback}")

try:
    example_function()

except ZeroDivisionError as e:
    print(f"Caught an exception: {e}")
````

## * Types of Plots available in Matplotlib *

The `Matplotlib` library in Python provides a wide range of visualization types to represent data in various ways. Here's a list of common visualizations you can create using `Matplotlib`:

### 1. **Line Plot (`plot`)**
   - **Description**: Plots data points connected by lines. Useful for showing trends over time or continuous data.
   - **Code**: `plt.plot(x, y)`

### 2. **Scatter Plot (`scatter`)**
   - **Description**: Displays data points (markers) on a 2D plane, typically used to show the relationship between two variables.
   - **Code**: `plt.scatter(x, y)`

### 3. **Bar Plot (`bar` or `barh`)**
   - **Description**: Displays rectangular bars to represent data. The height (or width for horizontal bars) of the bars represents the value.
   - **Code**: `plt.bar(x, height)` or `plt.barh(x, width)`

### 4. **Histogram (`hist`)**
   - **Description**: Shows the distribution of data by grouping data into bins and displaying the frequency of data points within each bin.
   - **Code**: `plt.hist(data, bins=number_of_bins)`

### 5. **Pie Chart (`pie`)**
   - **Description**: Divides a circle into proportional segments to represent different categories.
   - **Code**: `plt.pie(sizes, labels=categories)`

### 6. **Box Plot (or Box-and-Whisker Plot) (`boxplot`)**
   - **Description**: Shows the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
   - **Code**: `plt.boxplot(data)`

### 7. **Violin Plot (`violinplot`)**
   - **Description**: Similar to a box plot but also shows the probability density of the data at different values, giving a more detailed view of the data distribution.
   - **Code**: `plt.violinplot(data)`

### 8. **Area Plot (`fill_between`)**
   - **Description**: Displays a filled area under the line, useful for comparing volume over time.
   - **Code**: `plt.fill_between(x, y)`

### 9. **Error Bar Plot (`errorbar`)**
   - **Description**: Shows error margins in the form of vertical or horizontal error bars around each data point.
   - **Code**: `plt.errorbar(x, y, yerr=error_margin)`

### 10. **Heatmap (with `imshow` or `pcolormesh`)**
   - **Description**: Represents data in matrix form with color indicating the magnitude of values. Useful for showing correlations, matrices, or grid-based data.
   - **Code**: `plt.imshow(data, cmap='viridis')` or `plt.pcolormesh(data)`

### 11. **Contour Plot (`contour` and `contourf`)**
   - **Description**: Represents 3D data in two dimensions using contour lines (or filled contours) to show different levels.
   - **Code**: `plt.contour(X, Y, Z)` or `plt.contourf(X, Y, Z)`

### 12. **Quiver Plot (`quiver`)**
   - **Description**: Displays velocity vectors as arrows, typically used to represent wind flow, currents, or vector fields.
   - **Code**: `plt.quiver(X, Y, U, V)`

### 13. **Stem Plot (`stem`)**
   - **Description**: A discrete plot where each data point is represented by a vertical line extending from a baseline, ending with a marker.
   - **Code**: `plt.stem(x, y)`

### 14. **Polar Plot (`polar`)**
   - **Description**: A plot on a polar coordinate system, often used for circular data.
   - **Code**: `plt.polar(theta, r)`

### 15. **3D Plot (`plot_surface`, `plot_wireframe`, etc.)**
   - **Description**: Used to plot 3D data. Requires importing the `Axes3D` module from `mpl_toolkits.mplot3d`.
   - **Code**:
     - Surface Plot: `ax.plot_surface(X, Y, Z)`
     - Wireframe Plot: `ax.plot_wireframe(X, Y, Z)`
     - Scatter Plot in 3D: `ax.scatter(X, Y, Z)`

### 16. **Stacked Bar Plot**
   - **Description**: A bar plot where bars are stacked on top of each other to show cumulative data.
   - **Code**: `plt.bar(x, height1, bottom=height2)`

### 17. **Step Plot (`step`)**
   - **Description**: A line plot that uses steps to connect data points, useful for representing discrete changes.
   - **Code**: `plt.step(x, y)`

### 18. **Hexbin Plot (`hexbin`)**
   - **Description**: A 2D histogram plot where the data is divided into hexagonal bins, representing density.
   - **Code**: `plt.hexbin(x, y, gridsize=number_of_bins)`

### 19. **Event Plot (`eventplot`)**
   - **Description**: Visualizes events on a timeline, typically used for representing discrete occurrences or spikes.
   - **Code**: `plt.eventplot(data)`

### 20. **Logarithmic Plot (`semilogx`, `semilogy`, `loglog`)**
   - **Description**: Plots where one or both axes are logarithmic.
   - **Code**:
     - `plt.semilogx(x, y)` (log scale on x-axis)
     - `plt.semilogy(x, y)` (log scale on y-axis)
     - `plt.loglog(x, y)` (log scale on both axes)

### 21. **Density Plot (using `Seaborn` on top of `Matplotlib`)**
   - **Description**: A smoothed representation of the distribution of data.
   - **Code**: `sns.kdeplot(data)`

---

These are the most common types of visualizations available in `Matplotlib`. Each one serves different purposes depending on the data and the insights you want to highlight. You can customize them further with colors, labels, titles, annotations, and more.

In a **Matplotlib pie chart**, the `plt.pie()` function has several parameters that allow you to customize the charts appearance and behavior. Here's a detailed explanation of each parameter:

### 1. **`x`** (Required)
   - **Description**: The input data representing the sizes of the pie slices. This is typically a list, array, or similar structure that holds numerical values (e.g., `[30, 20, 50]`).
   - **Example**: `x=[30, 20, 50]`

### 2. **`labels`**
   - **Description**: A list of strings representing the labels corresponding to each slice. These labels are displayed alongside each wedge.
   - **Default**: `None`
   - **Example**: `labels=['A', 'B', 'C']`

### 3. **`colors`**
   - **Description**: A sequence of colors to be used for each slice of the pie. You can use a list of color names, hexadecimal color codes, or RGB tuples.
   - **Default**: Uses a default color cycle.
   - **Example**: `colors=['#ff9999', '#66b3ff', '#99ff99']`

### 4. **`autopct`**
   - **Description**: Controls how percentages are displayed on the slices. It accepts a string or a function.
     - **String format**: Example: `'%1.1f%%'` (to display percentages with 1 decimal place).
     - **Function**: Example: A lambda function to conditionally display percentages.
   - **Default**: `None` (percentages are not displayed).
   - **Example**: `autopct='%1.1f%%'` or `autopct=lambda p: '{:.0f}%'.format(p)`

### 5. **`explode`**
   - **Description**: A sequence of values to offset individual slices from the center. Each value represents how far the corresponding slice is "exploded" or pushed away from the center.
   - **Default**: `None`
   - **Example**: `explode=[0, 0.1, 0]` (the second slice will be "exploded" outward by 0.1 units).

### 6. **`startangle`**
   - **Description**: The starting angle of the pie chart, measured counterclockwise from the x-axis. Useful for rotating the pie chart to a specific orientation.
   - **Default**: `0` degrees.
   - **Example**: `startangle=90` (starts at the top of the circle).

### 7. **`shadow`**
   - **Description**: Adds a shadow behind the pie chart, giving it a 3D effect.
   - **Default**: `False`
   - **Example**: `shadow=True`

### 8. **`radius`**
   - **Description**: The radius of the pie chart. Controls the overall size of the pie.
   - **Default**: `1.0` (default unit circle).
   - **Example**: `radius=1.5` (increases the size of the pie).

### 9. **`counterclock`**
   - **Description**: Determines the direction in which the slices are plotted. If `True`, slices are plotted counterclockwise; if `False`, they are plotted clockwise.
   - **Default**: `True`
   - **Example**: `counterclock=False` (plots in a clockwise direction).

### 10. **`wedgeprops`**
   - **Description**: A dictionary to customize the appearance of the pie slices (wedges). You can specify properties like edge color, line width, etc.
   - **Default**: `None`
   - **Example**: `wedgeprops={'edgecolor': 'black', 'linewidth': 2}` (sets black edges and thicker lines).

### 11. **`textprops`**
   - **Description**: A dictionary to customize the appearance of text labels. You can set properties like font size, color, etc.
   - **Default**: `None`
   - **Example**: `textprops={'fontsize': 14, 'color': 'black'}` (sets the font size and color of the text labels).

### 12. **`center`**
   - **Description**: Specifies the center of the pie chart in (x, y) coordinates.
   - **Default**: `(0, 0)` (the center of the plot area).
   - **Example**: `center=(1, 1)`

### 13. **`frame`**
   - **Description**: When `True`, draws a frame around the chart.
   - **Default**: `False`
   - **Example**: `frame=True`

### 14. **`pctdistance`**
   - **Description**: The ratio between the center of the pie and the location of the percentage text. It controls how far the percentages are placed from the center.
   - **Default**: `0.6`
   - **Example**: `pctdistance=0.8` (moves the percentage text further away from the center).

### 15. **`labeldistance`**
   - **Description**: The radial distance at which the labels are drawn. The distance is measured as a fraction of the pie radius.
   - **Default**: `1.1`
   - **Example**: `labeldistance=1.2` (moves labels further away from the pie chart).

### 16. **`normalize`**
   - **Description**: Normalizes the values of the pie chart if the sum of the values is not equal to 1.
   - **Default**: `None`
   - **Example**: `normalize=True`

### 17. **`colors`**
   - **Description**: A sequence of colors for each wedge in the pie chart. It can be a list of colors, hex codes, or color names.
   - **Default**: `None` (uses the default Matplotlib color cycle).
   - **Example**: `colors=['#ff9999', '#66b3ff', '#99ff99']`

---

### Example Pie Chart with Multiple Parameters:

```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [30, 20, 50]
labels = ['A', 'B', 'C']
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = [0, 0.1, 0]  # Explode the second slice

# Create pie chart
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    explode=explode, 
    autopct='%1.1f%%', 
    startangle=90, 
    shadow=True,
    wedgeprops={'edgecolor': 'black'},
    textprops={'fontsize': 12}
)

plt.title('Customized Pie Chart')
plt.show()
```

This example demonstrates how to combine several parameters (`labels`, `colors`, `explode`, `autopct`, `startangle`, `shadow`, etc.) to create a fully customized pie chart.

### Types of plots available in Seaborn

Seaborn offers a wide variety of plots to help with statistical data visualization. Here are some of the common plot types in Seaborn:

### 1. **Relational Plots**
   - **scatterplot()**: Displays relationships between two variables as points.
   - **lineplot()**: Shows relationships between continuous data with lines.
   
### 2. **Categorical Plots**
   - **stripplot()**: A scatter plot that displays individual data points along a categorical axis.
   - **swarmplot()**: Similar to `stripplot()`, but the points are adjusted to avoid overlap, making it easier to see the distribution of points.
   - **boxplot()**: Displays the distribution of data based on five summary statistics: minimum, first quartile, median, third quartile, and maximum.
   - **violinplot()**: Combines a boxplot with a KDE (Kernel Density Estimate) plot to show data distribution.
   - **barplot()**: Shows the distribution of data with bars, useful for comparing categorical data with aggregate statistics.
   - **pointplot()**: Displays point estimates and confidence intervals for different categories.

### 3. **Distribution Plots**
   - **histplot()**: Displays the distribution of a dataset using a histogram.
   - **kdeplot()**: Shows the probability density function of a continuous variable using a smooth curve.
   - **distplot()** (Deprecated): Combines a histogram and a KDE plot.
   - **rugplot()**: Shows individual data points along the x-axis, often used with other plots.
   - **ecdfplot()**: Shows the empirical cumulative distribution function.

### 4. **Matrix Plots**
   - **heatmap()**: Displays data in a matrix format, useful for showing correlations or frequencies in a grid.
   - **clustermap()**: Similar to a heatmap but with hierarchical clustering to organize rows and columns.

### 5. **Pairwise Plots**
   - **pairplot()**: Creates a matrix of scatterplots or KDE plots for pairwise relationships between variables in a dataset.

### 6. **Facet Grids**
   - **FacetGrid**: A multi-plot grid for plotting conditional relationships across multiple subsets of data.
   - **lmplot()**: Plots linear regression models across different subsets of the dataset.
   - **catplot()**: Combines `FacetGrid` with categorical plots like `boxplot()`, `violinplot()`, or `barplot()`.

These plots can be customized extensively with Seaborn's intuitive API for color palettes, aesthetics, and annotations.
